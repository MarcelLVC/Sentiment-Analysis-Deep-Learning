```python
import os
os.environ["PATH"] += os.pathsep + 'C:/Program Files/Graphviz/bin'

from tensorflow.keras.utils import plot_model
import tensorflow as tf
```

```python
def plot_model(model):
  return tf.keras.utils.plot_model(
    model,
    to_file=str(model.name) + ".png",
    show_shapes=True,
    show_dtype=False,
    show_layer_names=True,
    rankdir="TB",
    expand_nested=True,
    dpi=96,
    layer_range=None,
)
```

```python
model_nn = tf.keras.models.load_model("nn_model.h5")
model_lstm_1 = tf.keras.models.load_model("lstm_model_1.h5")
model_lstm_2 = tf.keras.models.load_model("lstm_model_2.h5")
model_lstm_3 = tf.keras.models.load_model("lstm_model_3.h5")
```

```python
plot_model(model_nn)
```

```python
plot_model(model_lstm_1)
```

```python
plot_model(model_lstm_2)
```

```python
plot_model(model_lstm_3)
```
